```python
import tensorflow as tf
import numpy as np
from tensorflow import keras #keras is an API in tensorflow
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])

#Dense is used to defined the layer of connected neurons.
#There's only one dense here.
#So there's only one layer and there's only one unit in it, so it's a single neuron.
#Successive layers are defined in sequence, hence the word sequential.
#as, there's only one. So you have a single neuron.
#we define the shape of what's input to the neural network in the first and in this case the only layer, and you can see that our input shape is super simple. It's just one value.
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
#Optimizer = 'sgd' = stochastic gradient descent
#Loss function = 'mean squared error'
```

### providing the Data

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

### Training the Neural Network

```python
model.fit(xs,ys,epochs=500)
```

```python
print(model.predict([10.0]))
#the actual rule is "2*x - 1"
```
